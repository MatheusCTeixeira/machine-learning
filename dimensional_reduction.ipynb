```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
X, y = load_iris(return_X_y=True)
```

# Visualizing

Dimensinal Reduction (DR) plays an important role in Data Mining, since the "curse of dimensionality" can affect the performance of many algorithms. An alternative to (or complement) DR is Feature Selection (FS).  When used as a complement, a reasonable pipeline is: first apply FS, then DM, because the former can remove non-informative variables (using Variance Threshold) or features with correlated with the predicted variable (with Mutual Information). Now, with a subset of meaningful variables, we can reduce dimensionality keeping in mind the loss of information Ã— reduction of dimension trade off.

### PCA

```python
pca = PCA(n_components=2, random_state=1)
Xpca = pca.fit_transform(X)

df = pd.DataFrame(data={'x': Xpca[:, 0], 'y': Xpca[:, 1], 'label': y})
```

```python
sns.scatterplot(data=df, x='x', y='y', hue='label')
```

### t-SNE

```python
tsne = TSNE(n_components=2, perplexity=7, learning_rate=30, n_iter=1000, random_state=1)
Xtsne = tsne.fit_transform(X)

df = pd.DataFrame(data={'x': Xtsne[:, 0], 'y': Xtsne[:, 1], 'label': y})
```

```python
sns.scatterplot(data=df, x='x', y='y', hue='label')
```

### LDA

```python
lda = LinearDiscriminantAnalysis(n_components=2)
Xlda = lda.fit_transform(X, y)

df = pd.DataFrame(data={'x': Xlda[:, 0], 'y': Xlda[:, 1], 'label': y})
```

```python
sns.scatterplot(data=df, x='x', y='y', hue='label')
```
